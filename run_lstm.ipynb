```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
```

```python
mnist = input_data.read_data_sets('data', one_hot=True)

n_input = 28
max_time = 28
n_classes = 10
num_epoch = 5001

batch_size = 50
n_batch = mnist.train.num_examples // batch_size

lstm_size = 100

tf.reset_default_graph()
x = tf.placeholder(tf.float32,[None, 784])
y = tf.placeholder(tf.float32,[None, 10])

#tf.Variable 默认可训练变量
weights = tf.Variable(tf.truncated_normal([lstm_size, n_classes], stddev=0.1, dtype=tf.float32))
biases = tf.Variable(tf.constant(0.1, shape=[n_classes]))


def RNN(X, weights, biases):
    inputs = tf.reshape(X, [-1,28,28])
    
    with tf.name_scope('lsmt_cell'):
        lsmt_cell = tf.contrib.rnn.BasicLSTMCell(lstm_size)
    
    outputs, final_state = tf.nn.dynamic_rnn(lsmt_cell,inputs,dtype=tf.float32)
    results = tf.nn.softmax(tf.matmul(final_state[1], weights) + biases)
    
    return results

with tf.name_scope('RNN'): 
    prediction = RNN(x, weights, biases)

loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(labels=y, logits=prediction))

train_step = tf.train.AdamOptimizer(0.0001).minimize(loss)

correction_prediction = tf.equal(tf.argmax(y,1), tf.argmax(prediction,1))
accuracy = tf.reduce_mean(tf.cast(correction_prediction,dtype=tf.float32))

saver = tf.train.Saver()

init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    count = 0
    for epoch in range(50):
        for i in range(n_batch):
            count += 1
            xs_batch, ys_batch = mnist.train.next_batch(batch_size)
            sess.run(train_step, feed_dict={x:xs_batch,y:ys_batch})
            if count % 100 == 0:
                test_acc = sess.run(accuracy, feed_dict={x:mnist.test.images, y:mnist.test.labels})
                train_acc = sess.run(accuracy, feed_dict={x:mnist.train.images[:10000], y:mnist.train.labels[:10000]})
                print('Iter ' + str(count) + ' Training Accuracy=' + str(train_acc) +', Testing Accuracy=' + str(test_acc))
    saver.save(sess, 'lstm_net/my_net.ckpt')
```
